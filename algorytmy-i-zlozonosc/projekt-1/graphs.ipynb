```python
import pandas as pd 
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt 

sns.set_context("talk")
```

```python
results = "time-results-optimized-with-begin-end.csv"
results = pd.read_csv(results, sep=",")
# Filling options
random_fill = results.loc[results["Fill function"] == "Random"]
fill_100 = results.loc[results["Fill function"] == "Only 0 to 100 values"]
non_decreasing_fill = results.loc[results["Fill function"] == "Non-decreasing"]


plt.figure(figsize=(19, 12))
_ = sns.lineplot(data=random_fill, x="Array size", y="Time elapsed[s]", hue="Pivot function")
_ = plt.title("Wypełnianie tablicy elementami losowymi przekraczającymi zakres")
_ = plt.xlabel("Liczba elementów w tablicy")
_ = plt.ylabel("Czas sortowania[s]")
plt.savefig("ryciny/lineplot-random-fill.png")
plt.close()

plt.figure(figsize=(19, 12))
_ = sns.lineplot(data=fill_100, x="Array size", y="Time elapsed[s]", hue="Pivot function")
_ = plt.title("Wypełnianie tablicy elementami losowymi z zakresu od 0 do 100")
_ = plt.xlabel("Liczba elementów w tablicy")
_ = plt.ylabel("Czas sortowania[s]")
plt.savefig("ryciny/lineplot-100-fill.png")
plt.close()

plt.figure(figsize=(19, 12))
_ = sns.lineplot(data=non_decreasing_fill, x="Array size", y="Time elapsed[s]", hue="Pivot function")
_ = plt.title("Wypełnianie tablicy elementami niemalejącymi")
_ = plt.xlabel("Liczba elementów w tablicy")
_ = plt.ylabel("Czas sortowania[s]")
plt.savefig("ryciny/lineplot-nondecreasing-fill.png")
plt.close()

# Non decreasing fill to 100 000 size
non_decreasing_fill_to_100000 = non_decreasing_fill.loc[non_decreasing_fill["Array size"] <= 100000]
plt.figure(figsize=(19, 12))
_ = sns.lineplot(data=non_decreasing_fill_to_100000, x="Array size", y="Time elapsed[s]", hue="Pivot function")
_ = plt.title("Wypełnianie tablicy elementami niemalejącymi")
_ = plt.xlabel("Liczba elementów w tablicy")
_ = plt.ylabel("Czas sortowania[s]")
plt.savefig("ryciny/lineplot-nondecreasing-fill-to-100000-size.png")
plt.close()

# Non decreasing fill without end pivot
non_decreasing_fill_no_end = non_decreasing_fill.loc[non_decreasing_fill["Pivot function"] != "End"]
plt.figure(figsize=(19, 12))
_ = sns.lineplot(data=non_decreasing_fill_no_end, x="Array size", y="Time elapsed[s]", hue="Pivot function")
_ = plt.title("Wypełnianie tablicy elementami niemalejącymi")
_ = plt.xlabel("Liczba elementów w tablicy")
_ = plt.ylabel("Czas sortowania[s]")
plt.savefig("ryciny/lineplot-nondecreasing-fill-no-end.png")
plt.close()

# Non decreasing fill log scale
fig = plt.figure(figsize=(19, 12))
_ = sns.lineplot(data=non_decreasing_fill, x="Array size", y="Time elapsed[s]", hue="Pivot function")
_.set_yscale("log")
_ = plt.title("Wypełnianie tablicy elementami niemalejącymi")
_ = plt.xlabel("Liczba elementów w tablicy")
_ = plt.ylabel("Czas sortowania[s]")
plt.savefig("ryciny/lineplot-nondecreasing-fill-log-scale.png")
plt.close()
```
